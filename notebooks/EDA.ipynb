```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
TMDB_DATA = pd.read_csv("../data/raw/TMDB_all_movies.csv")
TMDB_DATA.head()
```

```python
TMDB_DATA.columns
```

```python
keep_cols = [
    'id', 'title', 'overview', 'tagline', 'genres',
    'cast', 'director', 'writers', 'music_composer',
    'vote_average', 'vote_count', 'imdb_rating', 'imdb_votes', 'popularity',
    'original_language', 'spoken_languages',
    'poster_path', 'release_date'
]

TMDB_DATA = TMDB_DATA[keep_cols]
TMDB_DATA.head()
```

```python
TMDB_DATA.shape
```

```python
TMDB_DATA.isnull().sum()
```

```python
TMDB_DATA.info()
```

```python
# Drop rows with missing title
TMDB_DATA = TMDB_DATA[TMDB_DATA['title'].notnull()]

# Fill numeric columns with 0
TMDB_DATA['vote_average'] = TMDB_DATA['vote_average'].fillna(0)
TMDB_DATA['vote_count'] = TMDB_DATA['vote_count'].fillna(0)
TMDB_DATA['popularity'] = TMDB_DATA['popularity'].fillna(0)
TMDB_DATA['imdb_rating'] = TMDB_DATA['imdb_rating'].fillna(0)
TMDB_DATA['imdb_votes'] = TMDB_DATA['imdb_votes'].fillna(0)

# Fill overview and tagline with empty string
TMDB_DATA['overview'] = TMDB_DATA['overview'].fillna('')
TMDB_DATA['tagline'] = TMDB_DATA['tagline'].fillna('')

# Fill selected text columns with 'unknown'
text_cols = ['original_language', 'director', 'genres', 'cast', 'spoken_languages', 'writers']
TMDB_DATA[text_cols] = TMDB_DATA[text_cols].fillna('unknown')

TMDB_DATA['poster_path'] = TMDB_DATA['poster_path'].fillna(np.nan)

TMDB_DATA = TMDB_DATA.drop(columns=['music_composer'])
```

```python
TMDB_DATA['release_date'] = pd.to_datetime(TMDB_DATA['release_date'], errors='coerce')
TMDB_DATA['release_year'] = TMDB_DATA['release_date'].dt.year
most_common_year = int(TMDB_DATA['release_year'].mode()[0])
TMDB_DATA['release_date'] = TMDB_DATA['release_date'].fillna(pd.to_datetime(f'{most_common_year}-01-01'))
TMDB_DATA.drop(columns=['release_year'], inplace=True)
```
